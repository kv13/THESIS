```python
import json
```

```python
filenames_dict = {}
```

```python
def find_filename(value):
    filename = ""
    words    = value.split("(")
    if len(words)>=2:
        parts    = words[0].split(".")
        filename = ".".join(parts[0:-1])
    
    return filename
```

```python
with open('data1.json') as json_file:
    data = json.load(json_file)
    
    for p in data:
        stack_trace = p['stack_trace']
        if stack_trace != []:
            stack_trace = stack_trace[0].split("at ")
            
            for count,value in enumerate(stack_trace):   
                filename = find_filename(value)
                if filename == "":
                    continue
                
                if filename in filenames_dict:
                    #find next file if exists
                    next_filename = ""
                    if count+1<len(stack_trace):
                        next_filename = find_filename(stack_trace[count+1])
                    if next_filename != "":
                        if next_filename not in filenames_dict[filename]:
                            filenames_dict[filename].append(next_filename)
                else:
                    #find next file if exists
                    next_filename = ""
                    if count+1<len(stack_trace):
                        next_filename = find_filename(stack_trace[count+1])
                    
                    filenames_dict[filename] = [next_filename]
```

```python
for key in filenames_dict:
    print("filename:",key,"connects with:",filenames_dict[key])
    print("\n")
```
